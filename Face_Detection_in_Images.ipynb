### Importing Libraries

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

### Loading the pre-trained Haar Cascade model for face detection

```python
face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

### Loading the image

```python
image_path = r'C:\Users\neeta_000\Desktop\Python\children.jpg'  # Replace with your image path
img = cv.imread(image_path)
```

```python
img
```

### Converting image to grayscale for face detection 

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

### Displaying the image using matplotlib

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

### Performing face detection

```python
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
```

```python
for (x, y, w, h) in faces:
        center = (x + w//2, y + h//2)
        radius = min(w, h) // 2
        cv.circle(img, center, radius, (0, 255, 0), 2)
```

### Display the result

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
