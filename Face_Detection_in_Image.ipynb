### Importing Libraries

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

### Loading the pre-trained Haar Cascade model for face detection

```python
face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

### Loading the image

```python
image_path = r'C:\Users\neeta_000\Desktop\Python\single_person.jpg'  
face1 = cv.imread(image_path)
```

```python
face1
```

### Converting image to grayscale for face detection

```python
gray = cv.cvtColor(face1, cv.COLOR_BGR2GRAY)
```

### Displaying the image using matplotlib

```python
plt.imshow(cv.cvtColor(face1, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

### Performing face detection

```python
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=10, minSize=(25, 40))
```

### Draw rectangles around the faces

```python
for (a, b, c, d) in faces:
    cv.rectangle(face1, (a, b), (a+c, b+d), (255, 0, 0), 2)
```

### Displaying the result

```python
plt.imshow(cv.cvtColor(face1, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
